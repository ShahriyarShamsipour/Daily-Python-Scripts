```python
pip install pymupdf

import fitz  # PyMuPDF

PDF_NAME = "Optimization_Algorithms_in_Machine_Learning_A_Meta_heuristics_Perspective.pdf"
PAGE_NUMBER = 1   # 1-based page number
DPI = 600         # increase to 1200 if needed

doc = fitz.open(PDF_NAME)

scale = DPI / 72
mat = fitz.Matrix(scale, scale)

page = doc.load_page(PAGE_NUMBER - 1)
pix = page.get_pixmap(matrix=mat, alpha=False)

output_name = f"{PAGE_NUMBER}.png"
pix.save(output_name)

doc.close()

print(f"Saved {output_name}")
```
